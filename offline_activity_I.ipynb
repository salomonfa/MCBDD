```python
import matplotlib.pyplot as plt
import seaborn as sns

def infection_prob(prevalence, specificity):
    if prevalence >= 0.001 and prevalence <= 50.0 and specificity in [99,99.9,99.99,99.999]:
        sensitivity = 0.99
        specificity = specificity / 100
        prevalence = prevalence / 100
        
        prob = (sensitivity * prevalence) / (sensitivity * prevalence + (1 - specificity) * prevalence)
        data_list = [prevalence * (1 - sensitivity), prevalence * sensitivity,
                    (1 - prevalence) * (1 - specificity),  (1 - prevalence) * specificity]
        label_list = ['infected / test neg.', 'infected / test pos.', 'not infected / test pos.', 'not infected / test neg.']
        colors = sns.color_palette('dark')
        explode = [0,0.1, 0,0]
        
        print('The probabilaty that Fred is infected is ' + str(prob))
        plt.pie(data_list, labels = label_list, colors = colors,explode = explode, autopct='%.2f%%', pctdistance=1.2, labeldistance=1.5)
        plt.show()
    else:
        print('prevalence must be between 0.001 and 50, specificity must be in [99, 99.9, 99.99, 99.999]')
```

```python
infection_prob(5, 99.9)
```
